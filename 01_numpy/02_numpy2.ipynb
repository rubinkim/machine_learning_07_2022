##### 그래프 그리는 경우 

```python
x = np.arange(-2, 2.005, 0.01)
x.size, x[0], x[-1]
```

```python
# linspace(시작, 끝, 몇등분)
x = np.linspace(-10, 10, 401)
x.size, x[-1]
```

```python
import matplotlib.pyplot as plt

y = np.sin(x)
z = np.cos(x)
plt.figure(figsize=(15, 5))
plt.plot(x, y)
plt.plot(x, z)
plt.grid()
plt.show()
```

##### Ramdom Sampling

```python
np.random.seed(seed=2021)
np.random.random()
```

```python
# 균등 분포
np.random.uniform(size=10)
```

```python
# 정규 분포 - N(1,2)
np.random.normal(1, 2, size=10)
```

```python
# 표준 정규 분포
stn_0 = np.random.normal(0, 1, size=10)
stn_1 = np.random.randn(10)
print(stn_0)
print(stn_1)
```

```python
rand_norm = np.random.randn(1000)
plt.figure(figsize=(15,5))
plt.hist(rand_norm, bins=30, color='green', edgecolor='yellow')  # Histogram
plt.show()
```
