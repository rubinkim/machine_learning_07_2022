#### The Poisson Distribution

##### The Poisson distribution models the probability of seeing a certain number of successes within a time interval, where the time it takes for the next success is modeled by an exponential distribution. It can be used to model traffic, such as the number of arrivals a hospital can expect in an hour or the number of emails I'd expect to receive in a week.

```python
import random
from scipy import stats
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
random.seed(12)

# Generate poisson data when average number of arrivals is 1.
arrival_rate_1 = stats.poisson.rvs(size=10000, mu=1)
print(pd.crosstab(index='counts', columns=arrival_rate_1))
```

```python
pd.DataFrame(data=arrival_rate_1).hist(bins=max(arrival_rate_1)+1, range=(-0.5, max(arrival_rate_1)+0.5),
                                       color='red', linewidth=2, edgecolor='yellow', figsize=(15, 8))
plt.show()
```
