#### Probability Distributions

##### A random variable is a variable that varies due to chance. Probability distribution describes how<br> random variables are distributed. There are a range of precisely defined probability distributions that<br> model different types of random variables.

##### Uniform Distribution<br>is a probability distribution where each value within a certain range is equally likely to occur. The scipy.stats library contains many useful fuctions for working with probility distributions.

##### stats.distributon.rvs() : generates random numbers from the specified distribution.<br>stats.distribution.cdf() : determines the probability that an observation drawn from a distribution falls below a specified value.<br>stats.distribution.ppf() : is the inverse of cdf(). It shows x axis cutoff value(quantile) associated with a given probability.<br>stats.distribution.pdf() : gives you the probability density(height of the distribution) at a given value.

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import random
from scipy import stats
```

```python
# 시작이 0이고, 끝이 10인 인터벌을 갖는 uniform분포에서 10000개의 data추출하기
random.seed(123)
uniform_data = stats.uniform.rvs(loc=0, scale=10, size=10000)
uniform_data
```

```python
uni_max, uni_min = np.max(uniform_data), np.min(uniform_data)
uni_mean, uni_std = np.mean(uniform_data), np.std(uniform_data)
print(f"uniform dist의 평균 : {uni_mean}\nuniform dist의 표준편차 : {uni_std}\n\
uniform dist의 최소값 : {uni_min}\nuniform dist의 최대값 : {uni_max}")
```

```python
pd.DataFrame(data=uniform_data).plot(kind='density', figsize=(15,6), xlim=(-2,12))
plt.show()
```

```python
cdf_5, cdf_95 =stats.uniform.cdf(x=[5,95], loc=0, scale=100)
cdf_5, cdf_95
```

```python
quantile_5, quantile_95 = stats.uniform.ppf(q=[0.05, 0.95], loc=0, scale=100)
quantile_5, quantile_95
```

```python
prob_5, prob_95 = stats.uniform.pdf(x=[0.05, 0.95], loc=0, scale=100)
prob_5, prob_95
```
