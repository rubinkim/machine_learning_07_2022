##### 라인 플롯

```python
plt.title('라인 플롯')
plt.plot([0, 1, 4, 9])
plt.grid()
plt.show()
```

```python
plt.title("라인 플롯")
xs = np.linspace(0, 3, 301)
ys = (xs + 1) ** 3
plt.plot(xs, ys)
plt.grid()
plt.show()
```

##### 스타일 지정

```python
# 색깔(r), 마커(s), 선스타일(--)
plt.title("'rs--' 스타일의 plot")
plt.plot([10,20,30,40],[1,4,9,16], 'rs--')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15, 5))
plt.plot([10,20,30,40],[1,4,9,16], c='b', lw=5, ls='--', marker='o', ms=15, mec='g', mew=5, mfc='r')
plt.grid()
plt.title("스타일 적용예", size=20, c='r')
plt.show()
```

##### 그림 범위 지정

```python
plt.figure(figsize=(15, 6))
plt.title("x축, y축의 범위 설정", size=30)
plt.plot([10,20,30,40],[1,4,9,16], c="b", lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.xlim(0,50)
plt.ylim(-2,20)
plt.grid()
plt.show()
```

##### 틱 설정

```python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("x축과 y축의 tick label 설정")
plt.plot(X, C)
plt.rc('axes', unicode_minus=False)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, +1])
plt.grid()
plt.show()
```

```python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("LaTex, 문자열로 tick label 정의")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1,0,1], ["Low", "Zero", "High"])
plt.grid()
plt.show()
```

##### 여러개의 선을 그리기

```python
t = np.arange(0, 5., 0.2)
plt.title("라인 플롯에서 여러개의 선 그리기")
plt.plot(t,t,'r--',  t,0.5*t**2,'bs:',  t,0.2*t**3,'g^-')
plt.grid()
plt.show()
```

##### 겹쳐 그리기

```python
plt.title("라인 플롯에서 여러개의 선 그리기")
plt.plot(t, t, 'r--')
plt.plot(t, 0.5*t**2, 'bs:')
plt.plot(t, 0.2*t**3, 'g^-')
plt.grid()
plt.show()
```
